```python
#image to vector conversion
import numpy as np 
import pandas as pd 
import os
import url_to_image

resolution = 224

import glob
import numpy as np
from PIL import Image

import torch
import torchvision.models as models
import torchvision.transforms as transforms
from torch.autograd import Variable
from PIL import Image

img_array = []

def get_resnet_vec(img_array):
    """
    This method expects an image in array format as input
    and returns a Pytorch tensor representing the image embedding
    """
    # Load pretrained model
    model = models.resnet18(pretrained=True)
    # Select the desired layer
    layer = model._modules.get('avgpool')
    model.eval()

    scaler = transforms.Resize((224, 224))
    normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                     std=[0.229, 0.224, 0.225])
    to_tensor = transforms.ToTensor()

    img = Image.fromarray(img_array)#.convert('RGB')  
    # Create a PyTorch Variable with the transformed image
    t_img = Variable(normalize(to_tensor(scaler(img))).unsqueeze(0))
    # Create a vector of zeros that will hold our feature vector
    # The 'avgpool' layer has an output size of 512
    my_embedding = torch.zeros(512)  # 4. Define a function that will copy the output of a layer

    def copy_data(m, i, o):
        my_embedding.copy_(o.data.reshape(o.data.size(1)))  # 5. Attach that function to our selected layer

    h = layer.register_forward_hook(copy_data)  # 6. Run the model on our transformed image
    model(t_img)  # 7. Detach our copy function from the layer
    h.remove()  # 8. Return the feature vector
    return my_embedding



files = glob.glob(r"C:\Users\Isha Tyagi\Desktop\OE Research (studentship)\open_images\*.jpg")
for my_file in files:
    print(my_file)
    
    image = Image.open(my_file).convert('RGB')
    image = np.array(image)
    if image is None or image.shape != (resolution, resolution, 3):
        print(f'This image is : {my_file} {image.shape if image is not None else "None"}')
        
    else:
        img_array.append(image)

        
embedding = get_resnet_vec(image)
print("\n")
print("\n")
print(embedding.size())
print(type(embedding))
print('img_array shape:', np.array(img_array).shape)
print("\n")

#verifying a torch tensor object is returned on first five values
for emb in embedding[:5] :
    print(emb)
    print(type(emb))
    print(emb.size())
    
    


    
 
```
